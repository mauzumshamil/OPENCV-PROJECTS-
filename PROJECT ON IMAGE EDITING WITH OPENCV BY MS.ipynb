```python
cv2.imshow('image',img)# pop up window name=image
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
img_grey=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
cv2.imshow('image',img_grey)# pop up window name=image
cv2.waitKey()
cv2.destroyAllWindows()  # we can zoom in  udandu and seen pixel numbers
```

```python
 # create binary image(black and white)--->only from gray scale image/not from rgb image--->first create gray scale then apply threshold
# 2 types conversion 1.manually give  each pixeland threshold give and said if pixel move greter than this threshold move to black other wise white
#this is not possible always
#2nd method automatic-->opencv study  that image and convert automatically 
(thresh,img_bin)=cv2.threshold(img_grey,128,255,cv2.THRESH_BINARY | cv2.THRESH_OTSU)  #OTSU-->automatic threshold algorithm(128 min,255(max))
```

```python
cv2.imshow('binary_image',img_bin)
cv2.waitKey()
cv2.destroyAllWindows() 
```

```python
#draw circle
img = cv2.imread(r"C:\Users\Lenovo\Downloads\img_1_opencv.jpg")
converted_image= cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
cir_img=cv2.circle(converted_image,center=(450,150),radius=100,color=(255,0,255),thickness=4)
plt.imshow(cir_img)

```

```python
#to draw filled shaped--->thickness=-ve value or filled
#draw circle
img = cv2.imread(r"C:\Users\Lenovo\Downloads\img_1_opencv.jpg")
converted_image= cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
cir_img=cv2.circle(converted_image,center=(450,150),radius=100,color=(100,0,0),thickness=cv2.FILLED)
plt.imshow(cir_img)
```

```python
#to write text
img = cv2.imread(r"C:\Users\Lenovo\Downloads\img_1_opencv.jpg")
converted_image= cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
text_img=cv2.putText(converted_image,text="hello",org=(50,300),fontFace=cv2.FONT_HERSHEY_COMPLEX,fontScale=3,color=(0,20,50),thickness=3)
#org=>text origin point,fontface-->style of font,fontscale-->size of font
plt.imshow(text_img)
```
