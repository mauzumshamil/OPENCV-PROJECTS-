## various image processing techniques and demonstrations

```python
import cv2 
import matplotlib.pyplot as plt
```

```python
from skimage import io
from PIL import Image
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
       "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
      "https://upload.wikimedia.org/wikipedia/commons/a/a5/Red_Kitten_01.jpg"]
```

```python
for url in urls:
    image = io.imread(url)
    plt.imshow(image)
    plt.axis('off')
    plt.show()
```

```python
for url in urls:
    image= io.imread(url)
    img_cvt = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
    final_img = cv2.hconcat((image,img_cvt))
    plt.imshow(final_img)
    plt.axis('off')
    plt.show()
    print('\n')
```

```python
print(image.dtype)
```

```python
plt.hist(image.ravel(),bins = 256, range= [0,256])
plt.show()
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
plt.hist(gray_image.ravel(),bins = 256, range = [0, 256])
plt.show()
```

# GRAY TO NEGATIVE

```python
im2 = 255 - gray_image  

plt.imshow(im2, cmap='gray')  
plt.axis('off')
plt.show()
```

#  intensity and brightness control

```python

im3 = (100.0 / 255) * gray_image + 100


plt.imshow(im3, cmap='gray')
plt.axis('off')  
plt.imshow(im3)
```

# a non-linear intensity transformation

```python
im4 = 255.0*(gray_image/255.0)**2

plt.imshow(im4, cmap='gray')
plt.axis('off')  
plt.show()
```
