# loading and presenting images directly from website

```python
urls = ["https://images.immediate.co.uk/production/volatile/sites/3/2020/09/Batman-Begins-f677ee0.jpg",
        "https://c.files.bbci.co.uk/15B5F/production/_131472988_531d444d-c439-4a68-bf4b-1a62c2abd31a.png",
        "https://static.dc.com/2023-03/dc_superman_hub_4up_Community_1x1.jpg?w=900"]
```

```python
for url in urls:
  image = io.imread(url)
  plt.imshow(image)
  plt.axis('off')
  plt.show()
  print('\n')
```

```python
io.imshow(image)
```

# changing BGR TO RGB

```python

for url in urls:
    image = io.imread(url)
    image2 = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
final_frame = cv2.hconcat((image,image2))
plt.imshow(final_frame)
plt.axis('off')
plt.show()
print('\n')
```

# histogram on image

A histogram is a graphical representation of the distribution of pixel values in an image.
 Creating a histogram with 256 bins (representing the intensity values from 0 to 255) to visualize the distribution of pixel intensities in the image variable.

```python
plt.hist(image.ravel(),bins=255,range=[0,255])
plt.show()
```
