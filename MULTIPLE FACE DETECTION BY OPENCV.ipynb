```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(r"C:\Users\dell\Downloads\multiface.jpg")
plt.imshow(img)
```

```python
cvrtd_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(cvrtd_img)
```

```python
face_cascade = cv2.CascadeClassifier(r"C:\Users\dell\Desktop\open cv\haarcascade_frontalface_default .xml")
```

```python
faces = face_cascade.detectMultiScale(cvrtd_img, scaleFactor=1.2, minNeighbors=5, minSize=(30, 30))
```

```python
len(faces)
```

```python
faces
```

```python
for (x,y,w,h) in faces:
    cv2.rectangle(cvrtd_img,(x,y),(x+w,y+h),color=(255,255,0),thickness=3)
plt.imshow(cvrtd_img)
```
