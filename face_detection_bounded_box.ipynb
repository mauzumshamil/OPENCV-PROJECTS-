```python
import cv2 
import matplotlib.pylab as plt
```

```python
img = cv2.imread(r"C:\Users\dell\Downloads\new.jpg")
cvrtd_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(cvrtd_img)
```

```python
face_cascade = cv2.CascadeClassifier(r"C:\Users\dell\Desktop\haarcascade_frontalface_default .xml")
```

```python
face_cascade.load(r"C:\Users\dell\Desktop\haarcascade_frontalface_default .xml")
```

```python
faces = face_cascade.detectMultiScale(cvrtd_img)
```

```python
faces
```

```python
len(faces)
```

```python
for (x,y,w,h) in faces:
    cv2.rectangle(cvrtd_img,(x,y),(x+w,y+h),color=(255,0,255),thickness=4)
plt.imshow(cvrtd_img)
plt.axis('off')
```
