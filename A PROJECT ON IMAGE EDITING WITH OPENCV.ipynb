```python
 # pop up window name=image
cv2.imshow('image',img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
img_grey = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img_grey)
plt.show()
```

```python
cv2.imshow('image',img_grey)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
(thresh,img_bin)=cv2.threshold(img_grey,128,255,cv2.THRESH_BINARY | cv2.THRESH_OTSU)
```

```python
cv2.imshow('binary_image',img_bin)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
plt.imshow(img_bin)
```

```python
org_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(org_img)
plt.show()
```
